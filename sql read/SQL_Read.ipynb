```python
#Dependencies
import pandas as pd
from sqlalchemy import create_engine
import requests
```

```python
#Extract CSV into DataFrame from movies
movies_file = "../OutputData/movies.csv"
movies_df = pd.read_csv(movies_file)
movies_df
```

```python
#setting the index to imdb_id
movies_df.set_index("imdb_id", inplace=True)

movies_df.head()

```

```python
#Extract CSV into DataFrame
platform_file = "../OutputData/platform.csv"
platform_df = pd.read_csv(platform_file)
platform_df
```

```python
#setting the index to platform_id
platform_df.set_index("platform_id", inplace=True)

platform_df.head()
```

```python
#Create Database connection
connection_string = "postgres:postgres@localhost:5432/movies_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
#Use pandas to Load  csv converted DataFrames into database
movies_df.to_sql(name='movieandshows', con=engine, if_exists='append', index=True)
```

```python
#confirm data has been added by quering the table
pd.read_sql_query('select * from moviesandshows', con=engine).head()
```

```python
#Use pandas to Load  csv converted DataFrames into database
platform_df.to_sql(name='source_table', con=engine, if_exists='append', index=True)
```

```python
#confirm data has been added by quering the table
pd.read_sql_query('select * from source_table', con=engine).head()
```
