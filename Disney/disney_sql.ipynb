```python
import requests
import json
import pandas as pd
from config import api_key
from pprint import pprint
```

```python
file = "Resources/disney_plus_shows.csv"

df = pd.read_csv(file)
df.head()
```

```python
# Dropping entire rows if NA value is there in imdb_id column

disney_movies = df.dropna(axis=0, subset=['imdb_id']).reset_index()
disney_movies.head()
```

```python
# Converting imdb_id column into list
imdb_list = disney_movies['imdb_id'].tolist()
imdb_list
```

```python
# Count of imdb id list
len(imdb_list)
```

```python
# Function to extract ratings for Rotten Tomatoes and Metacritic from Json response
def get_rating(ratings, site):
    for dic in ratings:
        if dic["Source"] == site:
            return dic["Value"]
```

```python
# Defining list
imdb_id = []
title = []
year = []
rated = []
released = []
runtime = []
genre = []
director = []
writer = []
actors = []
language = []
country = []
imdb_ratings = []
rt_ratings = []
metacritic_ratings = []
types = []

# For loop to iterate through each IMDB ID and parsing required data from JSON response
for ids in imdb_list:
    query_url = f"http://www.omdbapi.com/?apikey={api_key}&i={ids}"
    
    try:
        response = requests.get(query_url).json()
        
        imdb_id.append(response['imdbID'])
        title.append(response['Title'])
        year.append(response['Year'])
        rated.append(response['Rated'])
        released.append(response['Released'])
        runtime.append(response['Runtime'])
        genre.append(response['Genre'])
        director.append(response['Director'])
        writer.append(response['Writer'])
        actors.append(response['Actors'])
        language.append(response['Language'])
        country.append(response['Country'])
        imdb_ratings.append(response['imdbRating'])
        rt_ratings.append(get_rating(response['Ratings'], "Rotten Tomatoes"))
        metacritic_ratings.append(get_rating(response['Ratings'], "Metacritic"))
        types.append(response['Type'])
        
        print(f"Response recorded for movie with id {ids}")
        print("------------------------------")
    
    except KeyError:
        print(f"Movie with id {ids} was not found")
        print("------------------------------")
        pass
```

```python
# Defining dictionary to store JSON response 
disneyplus_db = pd.DataFrame({"IMDb ID": imdb_id, 
                              "Title": title, 
                              "Year": year,
                              "Rated": rated,
                              "Released": released,
                              "Runtime": runtime,
                              "Genre": genre,
                              "Director": director,
                              "Writer": writer,
                              "Actors": actors,
                              "Language": language,
                              "Country": country,
                              "IMDb Ratings": imdb_ratings,
                              "RT Ratings": rt_ratings, 
                              "Metacritic Ratings": metacritic_ratings, 
                              "Type": types})
disneyplus_db
```
