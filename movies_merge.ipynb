```python
# Dependencies
import pandas as pd
```

```python
# Converting clean Netflix csv to df
netflix_df = pd.read_csv("OutputData/NetflixDB.csv")
netflix_df
```

```python
# Converting clean Crackle csv to df
crackle_df = pd.read_csv("OutputData/crackle.csv")
crackle_df
```

```python
# Converting clean DisneyPlus csv to df
disneyplus_df = pd.read_csv("OutputData/disneyplus.csv")
disneyplus_df
```

```python
# Concat different csvs into single movies database and remove duplicates
frames = [netflix_df, crackle_df, disneyplus_df]
movies_db = pd.concat(frames).drop_duplicates(subset=["IMDB_id"], ignore_index=True)
```

```python
# Delete columns
del movies_db["RT_rating"]
del movies_db["Metacritic_rating"]
```

```python
# Display final movies database
movies_db
```

```python
# Convert movies dataframe to movies csv
movies_db.to_csv("OutputData/movies.csv", index=False)
```

```python
# Create new platform dataframe
platform_db = pd.DataFrame({"platform_id": ["P01", "P02", "P03"], 
                            "platform_name": ["Netflix", "Disney Plus", "Crackle"]})
platform_db
```

```python
# Convert platform dataframe to platform csv
platform_db.to_csv("OutputData/platform.csv", index=False)
```
